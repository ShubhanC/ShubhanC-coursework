## 4. Summarizing Data

We want to compare the sales of each video based on the publishing develop so we can understand which companies do the best when it comes to selling their video games and why. Since there are lots of publishing companies, we will take samples of 5 companies and compare their median sales for their video games. One sample will be from the top 50% of publishing companies (based on median sales) and the other sample will be from the bottom 50% of publishing companies.

We will do this to answer our question: **What is the relationship between video game publishing companies and their video game sales?**

```python
df_publish = df[["US Sales (millions)", "Publisher"]].groupby("Publisher").agg("median").reset_index()
```

```python
df_publish.head()
```

```python
df_publish.shape
```

```python
df_publish = df_publish.sort_values(by="US Sales (millions)", ascending=True)
```

```python
df_bottom, df_top = np.array_split(df_publish, 2)
```

```python
df_top.sample(5)
```

```python
df_bottom.sample(5)
```

```python
print(len(df_top), len(df_bottom))
```

```python
sample1 = df_top.sample(5)
sample1
```

```python
sample2 = df_bottom.sample(5)
sample2
```

```python
plt.bar(x=sample1["Publisher"], height=sample1["US Sales (millions)"])
plt.ylim(0, 1)
plt.xlabel("Publishing Company")
plt.ylabel("Sales (millions)")
plt.title("Sample of 5 high-performing Video Game Publishing Companies")
```

```python
plt.bar(x=sample2["Publisher"], height=sample2["US Sales (millions)"])
plt.ylim(0, 0.2)
plt.xlabel("Publishing Company")
plt.ylabel("Sales (millions)")
plt.title("Sample of 5 low-performing Video Game Publishing Companies")
```

```python
samples = pd.merge(sample1, sample2, how='outer')
samples
```

```python
plt.barh(y=samples["Publisher"], width=samples["US Sales (millions)"])
plt.xlim(0, 1)
plt.ylabel("Publishing Company")
plt.xlabel("Sales (millions)")
plt.title("All 10 Publishing Companies and their Video Game Sales")
plt.show()
```

```python
samples["US Sales (millions)"].describe()
```

We can see that in this sample distribution, the median video game sales is definitely skewed right. This can be shown by the high-end outlier of 0.86 million sales for Nintendo. The median is 0.107 million, and the IQR of this distribution is 0.13625.

Through this bar graph, we can conclude a few things. First, we found that it seems that only a few publishing companies are able to sell a lot. This was shown by the extreme outlier of Nintendo, which had a median 0.86 million sales, while the next highest in the sample had around 0.4 (Touchstone). Second, we realized that most smaller publishers seem to be competitive with each other, but they are unable to compete with Nintendo. This is shown through the IQR of 0.136 million, while the standard deviation was nearly 0.25.
